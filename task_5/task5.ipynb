````python
# Import Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os

# Set up matplotlib for displaying images
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
````

## 1. Reading & Writing Images

**Functions:**
- `cv2.imread()` - Read image from file
- `cv2.imwrite()` - Write image to file

**Parameters for imread:**
- Path to image
- Flag: 1 (color), 0 (grayscale), -1 (unchanged)

````python
# Example: Create a sample image for demonstration
# Creating a simple image with shapes
sample_image = np.zeros((300, 400, 3), dtype=np.uint8)
sample_image[50:150, 50:150] = [0, 255, 0]  # Green rectangle
sample_image[150:250, 200:350] = [255, 0, 0]  # Blue rectangle
sample_image[100:200, 100:300] = [0, 0, 255]  # Red rectangle

# Write image to file
output_path = "sample_image.jpg"
cv2.imwrite(output_path, sample_image)
print(f"Image saved to {output_path}")

# Read the image back
img = cv2.imread(output_path)
img_gray = cv2.imread(output_path, 0)  # Read as grayscale

print(f"Image shape (color): {img.shape}")
print(f"Image shape (grayscale): {img_gray.shape}")
print(f"Image dtype: {img.dtype}")
````

## 2. Displaying Images

**Functions:**
- `cv2.imshow()` - Display image in a window
- `cv2.waitKey()` - Wait for keyboard press (time in ms, 0 = infinite)
- `cv2.destroyAllWindows()` - Close all windows

**Note:** In Jupyter notebooks, use matplotlib instead for better display

````python
# Display images using matplotlib (recommended for Jupyter)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Convert BGR to RGB for proper color display
img_rgb = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)
axes[0].imshow(img_rgb)
axes[0].set_title("Color Image")
axes[0].axis('off')

# Display grayscale image
axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title("Grayscale Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()

# For standalone scripts, use cv2.imshow():
# cv2.imshow('Image Window', img)
# cv2.waitKey(0)  # Wait until user presses a key
# cv2.destroyAllWindows()  # Close the window
````

## 3. Resizing & Cropping Images

**Functions:**
- `cv2.resize()` - Resize image to specified dimensions
- Array slicing - Crop images using NumPy indexing

**Interpolation methods:**
- cv2.INTER_LINEAR (default)
- cv2.INTER_AREA (best for shrinking)
- cv2.INTER_CUBIC (smooth, slower)

````python
# Resize image to specific dimensions
resized = cv2.resize(img, (200, 150))
print(f"Original shape: {img.shape}")
print(f"Resized shape: {resized.shape}")

# Resize with specific interpolation
resized_area = cv2.resize(img, (150, 100), interpolation=cv2.INTER_AREA)

# Resize by scale factor
scale = 0.5
resized_scale = cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)
print(f"Resized by 0.5 scale: {resized_scale.shape}")

# Crop image using array slicing [y1:y2, x1:x2]
cropped = img[50:200, 50:250]
print(f"Cropped shape: {cropped.shape}")

# Visualize resizing and cropping
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title("Original Image")
axes[0, 0].axis('off')

axes[0, 1].imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
axes[0, 1].set_title("Resized (200x150)")
axes[0, 1].axis('off')

axes[1, 0].imshow(cv2.cvtColor(resized_scale, cv2.COLOR_BGR2RGB))
axes[1, 0].set_title("Resized by 0.5 scale")
axes[1, 0].axis('off')

axes[1, 1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axes[1, 1].set_title("Cropped Region")
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
````

## 4. Color Space Conversion

**Function:**
- `cv2.cvtColor()` - Convert image from one color space to another

**Common color space conversions:**
- cv2.COLOR_BGR2RGB - OpenCV uses BGR, matplotlib uses RGB
- cv2.COLOR_BGR2GRAY - Convert to grayscale
- cv2.COLOR_BGR2HSV - Convert to HSV (Hue, Saturation, Value)
- cv2.COLOR_BGR2YCrCb - Yb Cb Cr color space
- cv2.COLOR_BGR2LAB - LAB color space

````python
# Color space conversions
img_bgr = img.copy()
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
img_lab = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB)

# Visualize different color spaces
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title("RGB (Original)")
axes[0, 0].axis('off')

axes[0, 1].imshow(img_gray, cmap='gray')
axes[0, 1].set_title("Grayscale")
axes[0, 1].axis('off')

# HSV visualization (need to normalize for display)
axes[0, 2].imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
axes[0, 2].set_title("HSV")
axes[0, 2].axis('off')

# LAB visualization
lab_normalized = cv2.normalize(img_lab, None, 0, 255, cv2.NORM_MINMAX)
axes[1, 0].imshow(cv2.cvtColor(lab_normalized.astype(np.uint8), cv2.COLOR_LAB2RGB))
axes[1, 0].set_title("LAB")
axes[1, 0].axis('off')

# Show individual channels of HSV
axes[1, 1].imshow(img_hsv[:, :, 0], cmap='hsv')
axes[1, 1].set_title("HSV - Hue")
axes[1, 1].axis('off')

axes[1, 2].imshow(img_hsv[:, :, 1], cmap='gray')
axes[1, 2].set_title("HSV - Saturation")
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
````

## 6. Image Thresholding

**Functions:**
- `cv2.threshold()` - Simple binary thresholding
- `cv2.adaptiveThreshold()` - Adaptive thresholding (better for varying lighting)

**Thresholding Methods:**
- cv2.THRESH_BINARY - Pixels > threshold = 255, else = 0
- cv2.THRESH_BINARY_INV - Inverse binary
- cv2.THRESH_TRUNC - Pixels > threshold = threshold
- cv2.THRESH_TOZERO - Pixels < threshold = 0
- cv2.THRESH_TOZERO_INV - Inverse to zero

````python
# Image thresholding examples
# Create a gradient image for better demonstration
gradient = np.zeros((300, 400, 3), dtype=np.uint8)
for i in range(400):
    gradient[:, i] = [i * 255 // 400] * 3
gradient_gray = cv2.cvtColor(gradient, cv2.COLOR_BGR2GRAY)

# Simple binary threshold
ret, binary = cv2.threshold(gradient_gray, 127, 255, cv2.THRESH_BINARY)
print(f"Threshold return value: {ret}")

# Inverse binary
ret, binary_inv = cv2.threshold(gradient_gray, 127, 255, cv2.THRESH_BINARY_INV)

# Truncate
ret, trunc = cv2.threshold(gradient_gray, 127, 255, cv2.THRESH_TRUNC)

# Adaptive thresholding
adaptive = cv2.adaptiveThreshold(gradient_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY, 11, 5)

# Visualize thresholding methods
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].imshow(gradient_gray, cmap='gray')
axes[0, 0].set_title("Original Gradient")
axes[0, 0].axis('off')

axes[0, 1].imshow(binary, cmap='gray')
axes[0, 1].set_title("Binary Threshold")
axes[0, 1].axis('off')

axes[0, 2].imshow(binary_inv, cmap='gray')
axes[0, 2].set_title("Binary Inverse")
axes[0, 2].axis('off')

axes[1, 0].imshow(trunc, cmap='gray')
axes[1, 0].set_title("Threshold Truncate")
axes[1, 0].axis('off')

axes[1, 1].imshow(adaptive, cmap='gray')
axes[1, 1].set_title("Adaptive Threshold")
axes[1, 1].axis('off')

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
````

## 7. Edge Detection

**Function:**
- `cv2.Canny()` - Canny edge detection algorithm

**Parameters:**
- image: Input grayscale image
- threshold1: Lower threshold for the hysteresis process
- threshold2: Upper threshold for the hysteresis process
- apertureSize: Kernel size for Sobel operator (default 3)
- L2gradient: Use L2 norm for gradient magnitude (more accurate but slower)

````python
# Edge detection using Canny
# Use sample image converted to grayscale
gray = cv2.cvtColor(sample_image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection with different thresholds
edges_50_150 = cv2.Canny(gray, 50, 150)
edges_100_200 = cv2.Canny(gray, 100, 200)
edges_150_255 = cv2.Canny(gray, 150, 255)

# Visualize edge detection with different thresholds
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(gray, cmap='gray')
axes[0, 0].set_title("Original Grayscale")
axes[0, 0].axis('off')

axes[0, 1].imshow(edges_50_150, cmap='gray')
axes[0, 1].set_title("Canny Edges (50-150)")
axes[0, 1].axis('off')

axes[1, 0].imshow(edges_100_200, cmap='gray')
axes[1, 0].set_title("Canny Edges (100-200)")
axes[1, 0].axis('off')

axes[1, 1].imshow(edges_150_255, cmap='gray')
axes[1, 1].set_title("Canny Edges (150-255)")
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("Lower threshold values detect more edges (including noise)")
print("Higher threshold values detect only strong edges")
````

## 8. Contours and Shape Detection

**Functions:**
- `cv2.findContours()` - Find all contours in an image
- `cv2.drawContours()` - Draw contours on an image
- `cv2.contourArea()` - Calculate area of a contour
- `cv2.arcLength()` - Calculate perimeter of a contour
- `cv2.approxPolyDP()` - Approximate a contour with a polygon

**Parameters:**
- cv2.CHAIN_APPROX_SIMPLE - Simplify contours
- cv2.CHAIN_APPROX_NONE - Store all contour points

````python
# Create an image with clear shapes for contour detection
contour_image = np.zeros((400, 500, 3), dtype=np.uint8)

# Draw filled shapes
cv2.circle(contour_image, (150, 150), 60, (255, 255, 255), -1)
cv2.rectangle(contour_image, (250, 100), (400, 200), (255, 255, 255), -1)
cv2.circle(contour_image, (300, 300), 40, (255, 255, 255), -1)

# Convert to grayscale
contour_gray = cv2.cvtColor(contour_image, cv2.COLOR_BGR2GRAY)

# Find contours
contours, hierarchy = cv2.findContours(contour_gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print(f"Number of contours found: {len(contours)}")

# Analyze contours
for i, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    perimeter = cv2.arcLength(contour, True)
    print(f"Contour {i}: Area = {area:.2f}, Perimeter = {perimeter:.2f}")

# Draw contours on image
contour_result = cv2.cvtColor(contour_gray, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_result, contours, -1, (0, 255, 0), 2)

# Draw contours with different colors
contour_colored = contour_result.copy()
colors = [(0, 0, 255), (0, 255, 0), (255, 0, 0)]
for i, contour in enumerate(contours):
    color = colors[i % len(colors)]
    cv2.drawContours(contour_colored, [contour], 0, color, 2)
    # Add center point
    M = cv2.moments(contour)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
        cv2.circle(contour_colored, (cx, cy), 5, color, -1)

# Visualize contour detection
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image with Shapes")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(contour_result, cv2.COLOR_BGR2RGB))
axes[1].set_title("All Contours (Green)")
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(contour_colored, cv2.COLOR_BGR2RGB))
axes[2].set_title("Contours with Centers")
axes[2].axis('off')

plt.tight_layout()
plt.show()
````

## 9. Face Detection using Cascade Classifiers

**Function:**
- `cv2.CascadeClassifier()` - Load a pre-trained cascade classifier

**Pre-trained Cascade Classifiers Available:**
- haarcascade_frontalface_default.xml - Frontal face detection
- haarcascade_eye.xml - Eye detection
- haarcascade_smile.xml - Smile detection

**Method:**
- `detectMultiScale()` - Detect objects in the image

**Parameters:**
- scaleFactor: Image pyramid scale factor (1.1-1.4)
- minNeighbors: Detection threshold (5-7 recommended)
- flags: Detector flags
- minSize: Minimum object size
- maxSize: Maximum object size

````python
# Face Detection using Cascade Classifier
# Load the pre-trained cascade classifier
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_eye.xml')

print("Cascade classifiers loaded successfully")
print(f"Face cascade file: {cv2.data.haarcascades}haarcascade_frontalface_default.xml")

# Example: Face detection on a test image
# Note: For this demonstration, we'll create a synthetic image and show the detection method
def detect_faces_and_eyes(img):
    """Detect faces and eyes in an image"""
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )
    
    print(f"Faces detected: {len(faces)}")
    
    # Draw faces and detect eyes within faces
    img_with_detections = img.copy()
    
    for (x, y, w, h) in faces:
        # Draw rectangle around face
        cv2.rectangle(img_with_detections, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # Region of interest (face area)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img_with_detections[y:y+h, x:x+w]
        
        # Detect eyes in face
        eyes = eye_cascade.detectMultiScale(roi_gray)
        print(f"  Eyes detected in face: {len(eyes)}")
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
    
    return img_with_detections, faces

# Test with sample image (which doesn't have faces, but demonstrates the code)
result, detected_faces = detect_faces_and_eyes(sample_image)

# Visualize result
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
axes[1].set_title("Detected Faces (Blue) and Eyes (Green)")
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("\nNote: For real face detection, use an image containing actual faces!")
````

## Summary: Common OpenCV Workflows

### Image Processing Pipeline
1. **Read Image** → `cv2.imread()`
2. **Preprocess** → Resize, convert color space, denoise
3. **Feature Detection** → Edge detection, contours, corners
4. **Analysis** → Count objects, measure areas, classify shapes
5. **Output** → Draw annotations, write results

### Tips and Best Practices

1. **Color Space Handling:**
   - OpenCV uses BGR format, matplotlib uses RGB
   - Always convert: `cv2.cvtColor(img, cv2.COLOR_BGR2RGB)` for matplotlib

2. **Image Quality:**
   - Resize images for faster processing
   - Use appropriate thresholds for your use case
   - Test with different parameter values

3. **Performance:**
   - Use grayscale for faster processing when color isn't needed
   - Apply region of interest (ROI) for localized processing
   - Use cascade classifiers with appropriate parameters

4. **Debugging:**
   - Visualize intermediate results using matplotlib
   - Print shapes and data types of arrays
   - Test on sample images first

### Practical Applications
- **Medical Imaging:** Tumor detection, organ segmentation
- **Industrial Inspection:** Defect detection, quality control
- **Biometrics:** Face recognition, iris scanning
- **Autonomous Vehicles:** Lane detection, obstacle detection
- **Document Processing:** Text recognition, document scanning
- **Sports Analytics:** Player tracking, ball detection
